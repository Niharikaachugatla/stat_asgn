```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read csv file
hotel_data = pd.read_csv('hotel_booking.csv')
hotel_data
```

```python
hotel_data.info()
```

```python
# clean data and data preprocessing 
hotel_data.dropna()
```

```python
# drop duplicates
hotel_data.drop_duplicates()
```

```python
# basic statastics
description = hotel_data.describe()
description
```

```python
# correlation matrix
correlation_matrix = hotel_data.corr(numeric_only=True)
correlation_matrix 
```

```python
# Calculating statistical measures for ADR
adr_stats = {
    'Mean': hotel_data['adr'].mean(),
    'Median': hotel_data['adr'].median(),
    'Standard Deviation': hotel_data['adr'].std(),
    'Skewness': hotel_data['adr'].skew(),
    'Kurtosis': hotel_data['adr'].kurt()
}

# Displaying the results
for stat, value in adr_stats.items():
    print(f"{stat}: {value}")
```

```python
# Function to create a bar chart showing cancellation rate by hotel type
def plot_cancellation_rate_by_hotel(data):
    """
    Creates a bar chart showing the cancellation rate by hotel type.
    Args:
    - data (DataFrame): The hotel booking data.
    """
    plt.figure(figsize=(8, 6))
    cancellation_rate = data.groupby('hotel')['is_canceled'].mean() * 100
    cancellation_rate.plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title('Cancellation Rate by Hotel Type')
    plt.xlabel('Hotel Type')
    plt.ylabel('Cancellation Rate (%)')
    plt.xticks(rotation=0)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
plot_cancellation_rate_by_hotel(hotel_data)
```

```python
# Function to create a line plot for monthly booking trends
def plot_monthly_booking_trends(data):
    """
    Creates a line plot showing monthly booking trends over the years.
    Args:
    - data (DataFrame): The hotel booking data.
    """
    # Convert arrival_date_month to datetime format for proper ordering
    data['arrival_date'] = pd.to_datetime(data['arrival_date_year'].astype(str) + '-' + 
                                          data['arrival_date_month'] + '-01')
    monthly_counts = data.groupby(data['arrival_date'].dt.to_period('M')).size()
    plt.figure(figsize=(14, 6))
    monthly_counts.plot(kind='line', marker='o', linestyle='-', color='teal')
    plt.title('Monthly Booking Trends')
    plt.xlabel('Month')
    plt.ylabel('Number of Bookings')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
plot_monthly_booking_trends(hotel_data)
```

```python
# box plot adr distribution
def plot_adr_by_market_segment_clear(data):
    """
    Creates a box plot showing the ADR distribution across different market segments,
    with a y-axis limit set to 0-300 and reduced clutter for improved clarity.
    
    Args:
    - data (DataFrame): The hotel booking data.
    """
    plt.figure(figsize=(20, 12))
    sns.boxplot(data=data, x='market_segment', y='adr', palette='Pastel1', showfliers=False)
    # Using fewer data points and reducing alpha to make it less cluttered
    sns.stripplot(data=data, x='market_segment', y='adr', color='gray', alpha=0.1, jitter=0.1, dodge=True, size=2)
    plt.ylim(0, 300)  # Setting y-axis limit to 0-300 for better focus on typical ADR values
    plt.title('ADR Distribution by Market Segment (Zoomed to 0-300)')
    plt.xlabel('Market Segment')
    plt.ylabel('Average Daily Rate (ADR)')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
# Call the function to generate the zoomed-in box plot with y-axis from 0 to 300
plot_adr_by_market_segment_clear(hotel_data)
```
